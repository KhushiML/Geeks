# Binary Classification
Example 1: The 2d binary classification data generated by make_circles() have a spherical decision boundary.

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
```

```python
X, y = make_circles(n_samples = 200, shuffle = True, noise = 0.1, random_state = 42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

# Multi-Class Classification
Example 1: Data generated by the function make_blobs() are blobs that can be utilized for clustering.

```python
from sklearn.datasets import make_blobs
```

```python
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples = 500, centers = 3, n_features = 2, random_state = 23)
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

# Example 2: 
To generate data by the function make_classification() need to balance between n_informative, n_redundant and n_classes attributes X[:, :n_informative + n_redundant + n_repeated]

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples = 100, n_features = 2, n_redundant = 0, n_informative = 2, n_repeated = 0, n_classes = 3,
                          n_clusters_per_class = 1)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

# Example 3:
A random multi-label classification data is created by the function make make_multilabel_classification() 

```python
from sklearn.datasets import make_multilabel_classification
```

```python
import pandas as pd
```

```python
# Generate 2D Classification Dataset
```

```python
X,y = make_multilabel_classification(n_samples = 500, n_features =2, n_classes = 2, n_labels = 2, allow_unlabeled = True, 
                                    random_state = 23)
```

```python
# Create pandas dataframe fro generated datasets
df = pd.concat([pd.DataFrame(X, columns = ['X1','X2']), pd.DataFrame(y, columns = ["Label1","Label2"])],axis = 1)
display(df.head())
```

```python
# Plot the generated Datasets
plt.scatter(df['X1'],df['X2'], c=df['Label1'])
plt.show()
```

# Example 2:  
Multilabel feature using make_sparse_uncorrelated()

```python
from sklearn.datasets import make_sparse_uncorrelated
```

```python
# Generate 1d Regression Datasets
X, y = make_sparse_uncorrelated(n_samples = 100, n_features = 4, random_state =23)
```

```python
plt.figure(figsize = (8,6))
for i in range(4):
    plt.subplot(2,2, i+1)
    plt.scatter(X[:, i], y)
    plt.xlabel('X'+str(i+1))
    plt.ylabel('Y')
plt.show()
```
