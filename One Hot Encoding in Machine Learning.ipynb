# One Hot Encoding using Sci-kit Learn Library


```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df.dtypes
```

```python
df['Gender'] = df['Gender'].astype('category')
```

```python
df['Remarks'] = df['Remarks'].astype('category')
```

```python
#Assigning Numerical values and storing it in another columns
df['Gender_New'] = df['Gender'].cat.codes
df['Remarks_New'] = df['Remarks'].cat.codes
```

```python
df['Gender_New']
```

```python
# Create an instance of one-hot-encoder

enc = OneHotEncoder()
```

```python
# Passing Encoded Columns
enc_df = pd.DataFrame(enc.fit_transform(df[['Gender_New','Remarks_New']]).toarray())
```

```python
enc_df
```

```python
# Merge with main
new_df = df.join(enc_df)
```
